```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```
from sklearn.datasets import make_blobs # 등방성 가우시안 정규 분포를 이용, 가상데이터 생성
X, y = make_blobs(100, 2, centers = 2, random_state = 2, cluster_std = 1.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu');
```

```
# 전제 조건 ; 차원 사이에 공분산이 없는 가우스 분포를 따른다

# 모델 클래스 불러오기
from sklearn.naive_bayes import GaussianNB

# Model instance making
model = GaussianNB()

# 모델 적합 하기. Model training
model.fit(X, y)
```

```
# 새로운 데이터를 생성해서, Label을 예측
rng = np.random.RandomState(0) # 0 ~ 1 사이 난수 생성
Xnew = [-6, -14] + [14, 18] * rng.rand(2000, 2) # 2,000개 random 데이터 생성
# 각 차원에 14, 18 을 곱하고, -6, -14를 빼줌
# 1 번째 차원은 -6 ~ 8 사이
# 2 번째 차원은 -14 ~ 4 사이 숫자로 나옴
ynew = model.predict(Xnew)
```

```
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu');
lim = plt.axis()
plt.scatter(Xnew[:, 0], Xnew[:, 1], c=ynew, s=10, cmap='RdBu', alpha=0.2);
plt.axis(lim);
```

```
# 가우시안 나이브 베이즈
# 확률 표현의 장점
yprob = model.predict_proba(Xnew) # return probability estimates
yprob[- 10: ].round(2) # 마지막 에서 10개 표시
```
